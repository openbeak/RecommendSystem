```python
# import json module
import json
import csv

f = open('test.json')
json_data = json.load(f)
f.close()

# ranker 들의 푼 문제들 csv 파일로 저장
with open('test.csv', 'w', newline='') as file:
    # type : list -> dictionary
    for ranker in range(0,9):
        dic = json.loads(json_data[ranker])
    
        writer = csv.writer(file)
        writer.writerow(dic["solving_problem"])
```

```python
import numpy as np
import pandas as pd

f = open('test.csv', 'r')

lines = csv.reader(f)

# 0번 인덱스는 1000번 문제
# 문제번호 = 인덱스 + 1000
arr = np.zeros((9,18005))

ranker_num = 0
for line in lines:    
    for idx in range(0, len(line)):
        arr[ranker_num][int(line[idx])-1000] = 1
    ranker_num += 1
```

```python
'''df = pd.DataFrame(diabetes.data, columns=diabetes.feature_names)
df['target'] = diabetes.target
print(df.head())'''

df = pd.DataFrame(arr)
print(df.head())
```

```python
tmp_des = df.describe()
```

```python
# 정답률이 95% 이상이거나 5% 미만 문제는 제거
alive_col = np.zeros(18005)
num = 0
for idx in range(0, len(alive_col)):
    if tmp_des[idx]["mean"] >= 0.95:
        del df[idx]
    elif tmp_des[idx]["mean"] <= 0.05:
        del df[idx]
    else:
        alive_col[num] = idx
        num += 1
    
```
